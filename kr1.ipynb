```python
def password(length, speed, number_of_computer, power, cost):
    size = 62**length
    scientific_notation_size = "{:e}".format(size)
    print('1)—Ä–∞–∑–º–µ—Ä –º–Ω–æ–∂–µ—Å—Ç–≤–∞: 62^', length, "=",  scientific_notation_size)

    time = size/(speed * number_of_computer)
    day = 60 * 60 * 24
    days = time / day 
    scientific_notation = "{:e}".format(speed * number_of_computer)
    scientific_notation_time = "{:e}".format(time)
    scientific_notation_days = "{:e}".format(days)
    print('—Å—É–º–º–∞—Ä–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–æ–ª–µ–π/—Å: ', speed, '*', number_of_computer, '=', scientific_notation)
    print('—Å—É–º–º–∞—Ä–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ–∫—É–Ω–¥ –≤ –¥–Ω–µ: ', day)
    print('2)–Ω—É–∂–Ω–æ –≤—Ä–µ–º–µ–Ω–∏:', scientific_notation_size, '/', scientific_notation, '~~',  scientific_notation_time, 'c', ', –∞ –≤ –¥–Ω—è—Ö: ~~', days, '=', scientific_notation_days)
    
    answer = power * number_of_computer * time / 3600 * cost
    scientific_notation_answer = "{:e}".format(answer)

    print('3) –î–µ–Ω–µ–≥ –Ω—É–∂–Ω–æ: ', cost, '*', power, '*', number_of_computer, '*', time, '/ 3600 =', scientific_notation_answer, '—Ä—É–±')
```

```python
def I(x):
    return -np.log2(x)

def H(x1, x2):
    return -(x1 * np.log2(x1) + x2 * np.log2(x2))

def mutual_information(x1, x2, z1, z2):
    x1 = round(x1, 2)
    x2 = round(x2, 2)
    z1 = round(z1, 2)
    z2 = round(z2, 2)

    print('\n1) –°–æ–±—Å—Ç–≤–µ–Ω–Ω–∞—è –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è')
    print('I(x1) = ', round(I(x1), 2))
    print('I(x2) = ', round(I(x2), 2))

    print('\n2) –≠–Ω—Ç—Ä–æ–ø–∏—è –∏—Å—Ç–æ—á–Ω–∏–∫–∞ —Å–æ–æ–±—â–µ–Ω–∏–π, –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –∫–ª—é—á–µ–π –∏ —à–∏—Ñ—Ä–æ—Ç–µ–∫—Å—Ç–∞')
    print('H(x) = -[', x1, 'log2(', x1, ') + ', x2, 'log2(', x2, ')] = ', round(H(x1,x2), 2), ' –±–∏—Ç')
    print('H(z) = -[', z1, 'log2(', z1, ') + ', z2, 'log2(', z2, ')] = ', round(H(z1,z2), 2), ' –±–∏—Ç')
    y1 = round(x1*z1 + x2*z2, 2)
    y2 = round(1 - y1, 2)
    print('P(y=y1) = ', x1, '*', z1, ' + ', x2, '*', z2, '=', y1)
    print('P(y=y2) = 1-', y1, '=', y2)
    print('H(y) = -[', y1, 'log2(', y1, ') + ', y2, 'log2(', y2, ')] = ', round(H(y1,y2), 2),' –±–∏—Ç')

    print('\n3) –í–∑–∞–∏–º–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ—Ç–∫—Ä—ã—Ç–æ–≥–æ —Ç–µ–∫—Å—Ç–∞ –∏ –∫–ª—é—á–∞')
    print('H(x,z) = 0 —Ç–∞–∫ –∫–∞–∫ x –∏ z –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã')

    print('\n4) –í–∑–∞–∏–º–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ—Ç–∫—Ä—ã—Ç–æ–≥–æ —Ç–µ–∫—Å—Ç–∞ –∏ —à–∏—Ñ—Ä–æ—Ç–µ–∫—Å—Ç–∞')
    print('I(x,y) = H(y) - H(y|x) = H(y) - H(z) = ', round(H(y1,y2), 2), '-', round(H(z1,z2), 2), ' = ', round(H(y1,y2) - H(z1,z2), 2), ' –±–∏—Ç')
    print('—Ç–∞–∫ –∫–∞–∫ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º x –∑–Ω–∞—á–µ–Ω–∏–µ y –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –∫–ª—é—á–æ–º')

    print('\n5) –í–∑–∞–∏–º–Ω–∞—è –∏–Ω—Ñ–æ–æ—Ä–º–∞—Ü–∏—è –∫–ª—é—á–∞ –∏ —à–∏—Ñ—Ä–æ—Ç–µ–∫—Å—Ç–∞')
    print('I(z,y) = H(y) - H(y|z) = H(y) - H(x) = ', round(H(y1,y2), 2), '-', round(H(x1,x2), 2), ' = ', round(H(y1,y2) - H(x1,x2), 2), ' –±–∏—Ç')
    print('—Ç–∞–∫ –∫–∞–∫ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º y –∑–Ω–∞—á–µ–Ω–∏–µ x –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –∫–ª—é—á–æ–º')

    print('\n6)–ê–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –æ—Ç–∫—Ä—ã—Ç–æ–≥–æ —Ç–µ–∫—Å—Ç–∞')
    print('P(x1|y1) = ', x1, '*', z1, '/', y1, ' = ', round(x1*z1/y1, 2))
    print('P(x2|y1) = ', x2, '*', z2, '/', y1, ' = ', round(x2*z2/y1, 2))
    print('P(x2|y2) = ', x2, '*', z1, '/', y2, ' = ', round(x2*z1/y2, 2))
    print('P(x1|y1) = ', x1, '*', z2, '/', y2, ' = ', round(x1*z2/y2, 2))

    if z1 == z2:
        print('\n–ê–ø—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–º—É, —Å–∏—Å—Ç–µ–º–∞ –∞–±—Å–æ–ª—é—Ç–æ –Ω–∞–¥–µ–∂–Ω–∞')
    else:
        print('\n–ê–ø—Ä–∏–æ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–µ —Ä–∞–≤–Ω–æ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω–æ–º—É, —Å–∏—Å—Ç–µ–º–∞ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –∞–±—Å–æ–ª—é—Ç–Ω–æ –Ω–∞–¥–µ–∂–Ω–æ–π')
        print('–ß—Ç–æ–±—ã —Å–∏—Å—Ç–µ–º–∞ –±—ã–ª–∞ —É—Å—Ç–æ–π—á–∏–≤–∞ –Ω—É–∂–Ω–æ —á—Ç–æ–±—ã P(z=z1)=P(z=z2)=0.5')
        print('–ü–æ–∫–∞–∂–µ–º, —á—Ç–æ –∞–ø–æ—Å—Ç–µ—Ä–∏–æ—Ä–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ—Å–ª–µ –¥–æ—Ä–∞–±–æ—Ç–∫–∏ –±—É–¥—É—Ç —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º')
        print('—Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º –∞–±—Å–æ–ª—é—Ç–Ω–æ –Ω–∞–¥—ë–∂–Ω–æ–π –∫—Ä–∏–ø—Ç–æ—Å–∏—Å—Ç–µ–º—ã')
        z1 = 0.5
        z2 = 0.5
        y1 = x1*z1 + x2*z2
        y2 = 1 - y1
        print('P(y=y1) = '+str(x1)+'*'+str(z1)+str(x2)+'*'+str(z2)+'='+str(y1))
        print('P(y=y2) = 1-'+str(y1)+'='+str(y2))

        print('P(x1|y1) = ', x1, '*', z1, '/', y1, ' = ', x1*z1/y1, ' = P(x1) = ', x1)
        print('P(x2|y1) = ', x2, '*', z2, '/', y1, ' = ', x2*z2/y1, ' = P(x2) = ', x2)
        print('P(x2|y2) = ', x2, '*', z1, '/', y2, ' = ', x2*z1/y2, ' = P(x2) = ', x2)
        print('P(x1|y2) = ', x1, '*', z2, '/', y2, ' = ', x1*z2/y2, ' = P(x1) = ', x1)
```

```python
# b = a^-1 mod f -> ab = 1 * modf

def Euclidean_algorithm(basis, module):
  # step 0
  r = np.array(np.zeros(100));
  y = np.array(np.zeros(100));
  q = np.array(np.zeros(100));
  r[0] = module
  r[1] = basis

  y[0] = 0
  y[1] = 1

  print("r[-2] = ", r[0])
  print("r[-1] = ", r[1])
  print("y[-2] = ", y[0])
  print("y[-1] = ", y[1])

  # step 1
  # r[0] = r[1] * q[0] + r[2]

  q[0] = r[0] // r[1]
  r[2] = r[0] % r[1]
  y[2] = y[0] - q[0] * y[1]

  i = 2
  while r[i -1] != 1:
       r[i] = r[i - 2] % r[i - 1]
       q[i - 2] = r[i - 2] // r[i - 1]
       y[i] = y[i - 2] - q[i - 2] * y[i - 1]
       print("---------------------------------------------")
       print("q[", i - 2, "] = r[", i - 2, "] div r[", i - 1, "] = ", q[i - 2])
       print("r[", i - 2, "] = r[", i - 4, "] mod r[", i - 3, "] = ", r[i])
       print("y[", i - 2, "] = y[", i - 4, "] - q[", i - 4, "] * y[", i - 3, "] = ", y[i])
       i += 1
  b = y[i - 1]

  print("---------------------------------------------")
  print("The result of Euclidean algorithm is ", b)
  return b

  # input: module, 3 first elements
# x[n+1] = a * x[n] + b mod m
# x2 - x3 mod m = a((x1 - x2) mod m)
def find_elems(m, x1, x2, x3):
  print("x[n+1] = a * x[n] + b mod m")
  print("Solving the system of two equations we get")
  print("Coefficient a is:\na = (((", x2 - x3, ") mod m) * (", x1 - x2, ") ^ -1 mod m")

  a = (((x2 - x3) % m) * Euclidean_algorithm((x1 - x2) % m, m)) % m

  print("Coefficient a is:\na = (((", x2 - x3, ") mod m) * (", x1 - x2, ") ^ -1 mod m")

  print("---------------------------------------------")
  print("a = ", a)
  
  # x2 = x1 * a + b mod m
  # b mod m = (x2 - x1 * a) mod m

  print("b = (x2 - x1 * a mod m) mod m")
  b = (x2 - x1 * a % m) %m
  print("b = ", b)

  print("---------------------------------------------")

  print("x[n+1] = a * x[n] + b mod m")

  x4 = ((a * x3) %m  + b) % m
  x5 = ((a * x4) %m  + b) % m

  print("x4 = ", x4)
  print("x5 = ", x5)
```
